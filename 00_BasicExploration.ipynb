# Selecting and saving interesting data

```python
acidentes_ids = ocorrencia[ocorrencia.ocorrencia_classificacao == 'ACIDENTE'][['codigo_ocorrencia','codigo_ocorrencia1','codigo_ocorrencia2','codigo_ocorrencia3','codigo_ocorrencia4']]
acidentes_ids
```

```python
inst_aeronave = aeronave[inst_aeronave_cols]
inst_aeronave.head()
```

```python
inst_aeronave = inst_aeronave.drop_duplicates()
inst_aeronave.head()
```

```python
s =  set(acidentes_ids.codigo_ocorrencia2)
inst_aeronave['tipo_acidente'] = inst_aeronave.codigo_ocorrencia2.apply(lambda x: 'ACCIDENTE' if x in s else 'INCIDENTE')
```

```python
inst_aeronave['tipo_acidente'].value_counts()
```

```python
inst_aeronave.to_parquet(inst_path + "aeronave.parquet")
```
